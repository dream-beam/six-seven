We want to control the output power of a laser using a labjack T8

```python
from labjack import ljm

# Open LabJack T8
handle = ljm.openS("T8", "ANY", "ANY")

try:
    # Set DAC1 to 0.5V
    ljm.eWriteName(handle, "DAC1", 0.5)
    print("DAC1 set to 0.5V")
finally:
    # Close the connection
    ljm.close(handle)
```

```python
from labjack import ljm
from time import sleep
# Open LabJack T8
handle = ljm.openS("T8", "ANY", "ANY")

try:
    # Set DAC1 to 0.5V
    for i in range(5):
        ljm.eWriteName(handle, "DAC1", 0.01*(i+1))
        sleep(1)
    print("DAC1 set to 0.5V")
finally:
    # Close the connection
    ljm.close(handle)
```
