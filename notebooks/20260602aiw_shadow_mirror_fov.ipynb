We are studying the effect of the field of view on the beam size in the case of synchrotron optical

For the simulation, we consider a typical situation with an ellipsoidal mirror with an object distance p=15m and and an image distance q=5m, for a 2 deg grazing angle. 

The beam size is chosen to be 20 um-rms with a divergence of 50urad-rms, also quite typical.

Note that this code may be hard to run in VSCode (you need an old version of Shadow and intel libs) but can easily be run in Oasys 1.x python widget

## function linking beam size to field of view
beam size = F(relaive source position)

```python
def shadow_field_size(y_pos_m):

    beam = Shadow.Beam()
    oe0 = Shadow.Source()
    oe1 = Shadow.OE()
    oe2 = Shadow.OE()
    oe3 = Shadow.OE()

    # source
    oe0.FDISTR = 3
    oe0.F_COLOR = 2
    oe0.F_PHOT = 0
    oe0.F_POLAR = 0
    oe0.HDIV1 = 0.0
    oe0.HDIV2 = 0.0
    oe0.IDO_VX = 0
    oe0.IDO_VZ = 0
    oe0.IDO_X_S = 0
    oe0.IDO_Y_S = 0
    oe0.IDO_Z_S = 0
    oe0.ISTAR1 = 0
    oe0.NPOINT = 100000
    oe0.N_COLOR = 3
    oe0.PH1 = 799.9
    oe0.PH2 = 800.0
    oe0.PH3 = 800.1
    oe0.SIGDIX = 5e-05
    oe0.SIGDIZ = 0.0
    oe0.SIGMAX = 2e-05
    oe0.SIGMAZ = 0.0
    oe0.VDIV1 = 0.0
    oe0.VDIV2 = 0.0

    # dummy slits tro define source motion
    oe1.DUMMY = 100.0
    oe1.FSTAT = 1
    oe1.F_REFRAC = 2
    oe1.F_SCREEN = 1
    oe1.N_SCREEN = 1
    oe1.RDSOUR = 15.0
    oe1.T_IMAGE = 0.0
    oe1.T_INCIDENCE = 0.0
    oe1.T_REFLECTION = 180.0
    oe1.T_SOURCE = 0.0
    # source motion is here
    oe1.X_SOUR = y_pos_m
    oe1.Y_SOUR = 1e-09
    oe1.Z_SOUR = 1e-09

    # ellipsoidal mirror, cojugate parameters with internally calculated parameters
    oe2.ALPHA = 90.0
    oe2.DUMMY = 100.0
    oe2.FCYL = 1
    oe2.FMIRR = 2
    oe2.FWRITE = 1
    oe2.F_DEFAULT = 0
    #q
    oe2.SIMAG = 5.0
    #p
    oe2.SSOUR = 15.0
    # grazing angle
    oe2.THETA = 88.0
    oe2.T_IMAGE = 0.0
    oe2.T_SOURCE = 0.0

    oe3.DUMMY = 100.0
    oe3.F_REFRAC = 2
    oe3.F_SCREEN = 1
    oe3.N_SCREEN = 1
    oe3.T_IMAGE = 0.0
    oe3.T_INCIDENCE = 0.0
    oe3.T_REFLECTION = 180.0
    oe3.T_SOURCE = 5.0


    beam.genSource(oe0)
    beam.traceOE(oe1,1)
    beam.traceOE(oe2,2)
    beam.traceOE(oe3,3)

    x = beam.getshcol(1)
    y = beam.getshcol(3)

    #x_size_rms_m = numpy.std(x)

    #beam size
    y_size_rms_m = numpy.std(y)

    #Shadow.ShadowTools.plotxy(beam,1,3,nbins=101,nolost=1,title="Real space")
    # Shadow.ShadowTools.plotxy(beam,1,4,nbins=101,nolost=1,title="Phase space X")
    # Shadow.ShadowTools.plotxy(beam,3,6,nbins=101,nolost=1,title="Phase space Z")
    return y_size_rms_m
```

## compute and plot
(may take a while, typ 20 sec on my computer)

```python
import matplotlib.pyplot as plt 

source_pos_m = numpy.linspace(-0.010,0.010,101)
y_sizes_m = numpy.zeros(len(source_pos_m))
for i_s in range(len(source_pos_m)):
    y_size_rms_m = shadow_field_size(source_pos_m[i_s])
    y_sizes_m[i_s] = y_size_rms_m

plt.plot(source_pos_m*1e-3,y_sizes_m*1e6)
plt.xlabel("Source position (mm)")
plt.ylabel("Beam size (um)")
plt.title("Beam size at image plane as a function of source position")
plt.grid()
plt.show()
```
